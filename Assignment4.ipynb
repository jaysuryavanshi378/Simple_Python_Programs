###1. Generate a normal distribution and plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 0
std_dev = 1
size = 1000

data = np.random.normal(mean, std_dev, size)

plt.hist(data, bins=30,edgecolor ="black")
plt.title("Histogram of Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

###2. Compute mean, median, variance, and standard deviation for a dataset.

```python
data = [10, 20, 30, 40, 50]
n = len(data)

# Mean
mean = sum(data) / n

# Median
sorted_data = sorted(data)
if n % 2 == 0:
    median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
else:
    median = sorted_data[n//2]

# Variance (Population)
variance = sum((x - mean) ** 2 for x in data) / n

# Standard Deviation
std_dev = variance ** 0.5

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)

```

###3. Simulate a coin toss experiment and compute probabilities.

```python
import random

# Number of tosses
n = int(input("Enter number of coin tosses: "))

heads = 0
tails = 0

for i in range(n):
    toss = random.choice(['H', 'T'])
    if toss == 'H':
        heads += 1
    else:
        tails += 1

# Probabilities
prob_heads = heads / n
prob_tails = tails / n

print("Total Heads:", heads)
print("Total Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```
